```python
def column_clean_c_d(df):
    df.columns = np.append(df.iloc[0, 0:5], df.iloc[1, 5:])
    column_names = pd.Series(df.columns).str.replace("C\d\d: ", "")
    column_names = column_names.str.replace("D\d\d: ", "")
    column_names = column_names.str.replace(" \(.*\)", "")
    column_names = column_names.str.lower()
    column_names = column_names.str.strip()
    column_names = column_names.str.replace("care ", "")
    column_names = column_names.str.replace("for ", "") 
    column_names = column_names.str.replace("the ", "")
    column_names = column_names.str.replace(" – ", "_")
    column_names = column_names.str.replace(" - ", "_")
    column_names = column_names.str.replace(" ", "_")
    column_names = column_names.str.replace("–", "_")
    column_names = column_names.str.replace("-", "_")
    return column_names
```

```python
def drop_columns(df): 
    dropped_columns = []
    for i in df:
        if np.sum(df[i].isnull()) == df.shape[0]:
            dropped_columns.append(i)
    df = df.drop(dropped_columns, axis = 1)
    return df
```

```python
def column_clean(df): 
    df.columns = np.append(df.iloc[0, 0:4], df.iloc[1, 4:])
    column_names = pd.Series(df.columns).str.replace(" - (DMC\d\d)", "")
    column_names = column_names.str.replace(" - (DMD\d\d)", "")
    column_names = column_names.str.replace(" - (DME\d\d)", "")
    column_names = column_names.str.split(" - ").apply(lambda x: x[-1])
    column_names = column_names.str.split(" – ").apply(lambda x: x[-1])
    column_names = column_names.str.replace(" \(.*\)", "")
    column_names = column_names.str.lower()
    column_names = column_names.str.replace(" ", "_")
    column_names = column_names.str.replace("-", "_")
    column_names = column_names.str.replace("'", "")
    return column_names

def column_clean2(df): 
    df.columns = df.iloc[0, :]
    column_names = pd.Series(df.columns).str.replace("(HD\d)", "")
    column_names = column_names.str.replace("(DD\d)", "")
    column_names = column_names.str.replace(" \(.*\)", "")
    column_names = column_names.str.replace(":", "")
    column_names = column_names.str.replace(",", "")
    column_names = column_names.str.lower()
    column_names = column_names.str.replace("'", "")
    column_names = column_names.str.strip()
    column_names = column_names.str.replace(" ", "_")
    column_names = column_names.str.replace("-", "_")
    return column_names
```

```python
def clean(data):
    #Figuring out what are unique text entries to signify no data
    text = []
    for i in data.columns[2:]: 
        for j in data[data[i].str.contains("[A-Za-z]")][i].unique():
            if j not in text:
                text.append(j)

    #Dropping columns that have no data 
    data.index = np.arange(data.shape[0])
    for i in data.columns:
        if np.sum(np.array([data[i][j] in text for j in range(data.shape[0])])) == data.shape[0]:
            data = data.drop(i, axis = 1)
    
    #Dropping rows that have no data
    rows = []
    for i in range(data.shape[0]):
        if np.sum(np.array([data.iloc[i, j] not in text for j in range(1, data.shape[1])])) == 0:
            rows.append(i)
    data = data.drop(rows, axis = 0)
    
    #Getting rid of % sign 
    data = data.apply(lambda x: x.str.split("%").apply(lambda y: y[0]), axis = 0)
    
    #Converting no data to -1 
    ids = data["id"]
    name = data["organization_marketing_name"]
    data = data.drop(["id", "organization_marketing_name"], axis = 1)
    data = data.apply(lambda x: x.apply(lambda y: -1 if y in text else float(y)), axis = 0)
    data["id"] = ids
    data["name"] = name
    return data

data15 = clean(data15)
data16 = clean(data16)
data17 = clean(data17)
data18 = clean(data18)
data19 = clean(data19)
data15_2 = clean(data15_2)
data16_2 = clean(data16_2)
data17_2 = clean(data17_2)
data18_2 = clean(data18_2)
data19_2 = clean(data19_2)

data15_star = clean(data15_star)
data16_star = clean(data16_star)
data17_star = clean(data17_star)
data18_star = clean(data18_star)
data15_2_stars = clean(data15_2_stars)
data16_2_stars = clean(data16_2_stars)
data17_2_stars = clean(data17_2_stars)
data18_2_stars = clean(data18_2_stars)
```
